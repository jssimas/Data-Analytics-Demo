```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
velocidade = (3, 4, 6, 7)
tempo = (20, 30, 60, 100)
plt.plot(velocidade, tempo)
plt.grid()
plt.title("Velocidade pelo tempo") 
plt.xlabel("tempo")
plt.ylabel("velocidade")
plt.show()
plt.bar(velocidade, tempo)
plt.show()
```

```python

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel = "tempo (s)", ylabel = "tens√£o (mV)",
       title = "Tens√£o pelo tempo")
ax.grid()

# fig.savefig("test.png")
plt.show()
```

```python

t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_yticks(np.arange(-3, 3, 0.5))
ax.set_xticks(np.arange(-3, 3, 0.5))

ax.set(xlabel = "tempo (s)", ylabel = "tens√£o (mV)", title = "Tens√£o pelo tempo")
ax.grid()

# fig.savefig("test.png")
plt.show()
```
